Interpolation is commonly used to fill missing values in a table or a dataset using the already known values.<br>
Numpy provides the `interp1d()` to do linear or cubic interpolation.

```python
import numpy as np
```

Travel class and fare

```python
class_travel = [1, 3, 4]
fare_travel = [100, 60, 20]
```

Interpolate the fare for the 2nd class 

```python
np.interp(2, class_travel, fare_travel)
```

The `class_travel` is treated as an index here, and based on these values the `fare_travel` is interpolated.<br>
 **Note**: Make sure the data ius sorted based on the index array before you interolate.
 image.png

What if there were intermediate classes?

```python
np.interp([1.5, 2, 2.5, 2.9], class_travel, fare_travel)
```

**Sine Wave Interpolation**

```python
# turn-off scientific notation
np.set_printoptions(suppress=True)
```

First generate 25 points of sine wave. then interpolate the intermediate points linarly.

```python
x = np.linspace(0, 4*np.pi, 25)
y = np.sin(x)
```

```python
# plot
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(x, y, 'ro-')
plt.show()
```

```python
# interpolate
x_vals = np.linspace(0, 4*np.pi, 100)
y_interp = np.interp(x_vals, x, y)
y_interp
```

```python
# plot
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(x, y, 'ro-')
plt.plot(x_vals, y_interp, '-x')
plt.show()
```

**Interpolate with SciPy**

```python
from scipy import interpolate
```

```python
fx = interpolate.interp1d(x, y)
fx
```

```python
x_vals = np.linspace(0, 4*np.pi, 100)
y_new = fx(x_vals)
y_new
```

```python
# plot
plt.plot(x, y, 'ro')
plt.plot(x_vals, y_new, '-x')
plt.show()
```

**Cubic interpolation is more appropriate in this case**

```python
fx = interpolate.interp1d(x, y, kind='cubic')
fx
```

```python
y_new = fx(x_vals)
y_new
```

```python
# more smoother curve
plt.show()
```

```python
# 1. Solution
import numpy as np
arr = np.genfromtxt("Datasets/Class_and_Fare.csv", delimiter = ',', skip_header = 1)
arr[:10]
```

```python
# create the inputs to np.interp
x = arr[np.isnan(arr[:, 1]), 0]
xs = arr[~np.isnan(arr[:, 1]), 0]
ys = arr[~np.isnan(arr[:, 1]), 1]
```

```python
# sort x and y before interpolating
loc = np.argsort(xs)
xs = xs[loc]
ys = ys[loc]
```

```python
# now interpolate
y = np.interp(x, xs, ys)
y
```

```python
# result
np.c_[x, y]
```

```python
# # just to visualize
# plt.plot(xs, ys, 'ro')
# plt.plot(x, y, '-x')
# plt.show()
```

```python
# 2. Solution
from scipy import interpolate
x = np.linspace(-18, 18, 36)
noise = 0.1* np.random.random(len(x))
signal = np.sinc(x) + noise

# input
x_inp = np.linspace(-18, 18, 180)
```

```python
# linear
model_linear = interpolate.interp1d(x, signal, kind='linear')
y_linear = model_linear(x_inp)
y_linear[:5]
```

```python
# plot of linear interpolation
plt.plot(x, signal, 'ro')
plt.plot(x_inp, y_linear, '-x')
plt.show()
```

```python
# cubic interpolation
model_cubic = interpolate.interp1d(x, signal, kind='cubic')
y_cubic = model_cubic(x_inp)
y_cubic[:5]
```

```python
# plot of cubic interpolation
plt.plot(x, signal, 'ro')
plt.plot(x_inp, y_cubic, '-x')
plt.show()
```

**Plot**

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# all in one visualization
plt.figure(figsize=(12, 7))
plt.plot(x, signal, 'o', label='data')
plt.plot(x_inp, y_linear, 'r--', label='linear')
plt.plot(x_inp, y_cubic, '-', label='cubic')

plt.title('Interpolation: Cubic vs Linear', fontdict={'size': 16})
plt.legend()
plt.show()
```
