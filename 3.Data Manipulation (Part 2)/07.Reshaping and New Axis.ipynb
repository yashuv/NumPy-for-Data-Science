```python
import numpy as np
```

**Create 3x2 array**

```python
arr = np.array([[1, 2], [3, 4], [5, 6]])
arr
```

```python
arr.shape
```

**Reshape it to 2X3 form**

```python
arr2 = arr.reshape(2, 3)
arr2
```

```python
print(id(arr))
print(id(arr2))
```

**If we don't know the number of rows but know the column before hand, set n_rows=-1**

```python
arr = np.arange(69)
arr
```

```python
arr.reshape(-1, 3)
```

**`resize()` changes the shape of array in place. No need to reassign**

```python
arr = np.arange(15)
arr
```

```python
id(arr)
```

```python
arr.resize(5, 3)
```

```python
arr
```

```python
id(arr)
```

**Add an axis: column or row**

```python
arr = np.arange(5)
arr
```

```python
# make it as a column of a 2d array
arr1 = arr[:, np.newaxis]
arr1
```

```python
arr1.shape
```

```python
# make it as a row of 2d array
arr2 = arr[np.newaxis, :]
arr2
```

```python
arr2.shape
```

**Flatten an array**<br>
Flattens an array by creating a new copy of it

```python
arr = np.array([[1, 2],[3, 4],[5, 6]])
arr
```

```python
a2 = arr.flatten()
a2
```

```python
print(id(arr))
print(id(a2))
```

**Ravel**<br>
Ravel doesn't create a new array. It is a flattened view of the original one.

```python
a3 = arr.ravel()
```

```python
id(arr)  # same id as arr. so same as 'arr'
```

```python
id(a3)
```

```python
a3
```

```python
arr
```

Use `arr.flat()` to iterate
