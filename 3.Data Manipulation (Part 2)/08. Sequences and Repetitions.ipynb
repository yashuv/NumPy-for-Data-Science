**Create a simple sequence**

```python
import numpy as np
a = np.arange(10)
a
```

**Sequence with steps**

```python
# steps of 3
np.arange(1, 20, 3)
```

Reverse direction with negative step

```python
np.arange(20, 1, -3)
```

**When we don't know the step size**<br>
Use `np.linspace` Linear and equally sapces sequence.

```python
np.linspace(1, 2, 10) # here we know no. of items, i.e: 10
```

To see the stepsize used, set `retstep=True`

```python
np.linspace(1, 2, 10, retstep=True)
```

**Log Space**<br>
Similar to `linspace` but in log

```python
# log values: log(10) = 1 to log(100) = 2
np.logspace(1, 2, 10)
```

**Repeating Sequences** <br>
`np.tile(x, n)` will repeat sequence `x`, `n` times

```python
np.tile([1,2,3,4], 3)
```

Repeat each item n times

```python
np.repeat([1,2,3,4], 3)
```

**Practise challenge**<br>
Calculate the following pattern using the following input, without typing the  sequence itself. <br>
**Input**:<br>
a = np.array([1,2,3])<br>
**Output**:<br>
array([1, 1, 1, 2, 2, 2, 3, 3, 3, 1, 2, 3, 1, 2, 3, 1, 2, 3])

```python
a = np.array([1,2,3])

# solution 1:*
output = np.concatenate((np.repeat(a, 3) , np.tile(a, 3)))
output                             
```

```python
# solution 2:
np.array([np.repeat(a, 3), np.tile(a, 3)]).flatten()
```
