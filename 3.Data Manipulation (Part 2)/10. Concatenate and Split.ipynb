**Create two 2D arrays**

```python
import numpy as np
a = np.arange(12).reshape(3, 4)
a
```

```python
b = a * 2
b
```

**Concatenate Horizontally**<br>
I. `hstack` <br>
Takes only one argument

```python
np.hstack([a, b])
```

II. `c_`

```python
np.c_[a, b]
```

III. `column_stack` <br>
Takes only one argument

```python
np.column_stack([a, b])
```

```python
np.concatenate([a, b], axis=1)
```

**Concatenate Vertically**

```python
np.vstack([a, b])
```

```python
np.r_[a, b]
```

```python
np.row_stack([a, b])
```

```python
np.concatenate([a, b], axis=0)
```

**Splitting an array**

```python
arr = np.concatenate([a, b, a, b])
arr
```

Cut horizontally into 2 equal pieces

```python
# Split vertically (row-wise) into 2 pieces (horizontal cut)
np.vsplit(arr, 2)
```

Cut horizontally into 3 pieces at 3rd and 7th positions.

```python
np.vsplit(arr, [3, 7])
```

```python
# Split horizontally into 2 pieces (vertical cut)
np.hsplit(arr, 2)
```
