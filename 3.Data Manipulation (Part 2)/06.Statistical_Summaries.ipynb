```python
# I : fill in the missing values with mean
arr_miss[np.isnan(arr_miss)] = np.nanmean(arr_miss).round(2)
arr_miss
```

```python
# we can see that filled mean value is far from original value
print(arr_miss[missing_location])
print(arr[missing_location])
```

**Re-run `arr_miss` creation again**

```python
# II. fill the missing values with the mean of front and back value
for pos, val in enumerate(arr_miss):
  if np.isnan(val):
    arr_miss[pos] = np.nanmean([arr_miss[pos-1], arr_miss[pos+1]]).round(2)
```

**Practise Challenge 2**:<br>
Given the following array:<br>
I. Find the interquartile range<br>
II. Cap the values at 5%ile and 95%ile. Replace values greater than 95%ile value with 95%ile value.<br>

**create normal array as below**:<br>
np.random.seed(100)<br>
arr = np.random.normal(100, 75, size = 300).round(2)<br>
arr

```python
import numpy as np

#turn off scientific (e+02) notations
np.set_printoptions(suppress=True)

# create normal
np.random.seed(100)
arr = np.random.normal(100, 55, size=200).round(2)
arr
```

**Solution**

```python
# I. interquartile range
q25 = np.quantile(arr, 0.25)
q75 = np.quantile(arr, 0.75)
IQR = q75 - q25
print("IQR = ", IQR)
```

```python
# II. capping

# compute percentiles
q5 = np.quantile(arr, 0.05)
q95 = np.quantile(arr, 0.95)
print(q5)
print(q95)
```

```python
# cap it
arr = np.where(arr < q5, q5, arr)
arr = np.where(arr > q95, q95, arr)

arr
```
