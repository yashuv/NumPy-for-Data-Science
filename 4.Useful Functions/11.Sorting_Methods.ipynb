**Now working with 2D array data, with no dtype**

```python
data = np.genfromtxt("Datasets/Lifecyclesavings.csv", delimiter=',', skip_header=1)
data[:10]
```

```python
data.shape
```

Trying to sort a 2D array will sort every column of that array. As a result, the row contents get disturbed. Here we should be careful about data integrity!!

```python
data.sort(axis=0)  
data[:10]
# here each row values are changed (different from original). hence corrupted. 
```

So use `argsort` instead by importing  the original data again

```python
data = np.genfromtxt("Datasets/Lifecyclesavings.csv", delimiter=',', skip_header=1)
data[:10]
```

`np.argsort` gives the row positions of the numbers in increasing order

```python
# get row positions of increasing values
x = np.argsort(data[:, 0])
x
```

```python
# sort the data
data[x, :][:10]
```

Reverse sort

```python
# in reverse 
data[x[::-1], :][:10]
```

**Lexsort**<br>
if we want to sort based on more than one column, use `lexsort`.<br>
Let's first round the data, so some values will repeat

```python
data = data.round()
data[:10]
```

**Sort by 0th column first then by 1st column**<br>
Note the order of arguments.

```python
x = np.lexsort((data[:, 1].tolist(), data[:, 0].tolist()))
x
```

```python
data[x, :][:10]
```

**Practice Challenge**<br>
Sort 2nd column in ascending order and 4th column in descending order.

```python
import numpy as np
np.set_printoptions(suppress=True)

data = np.genfromtxt("Datasets/Lifecyclesavings.csv", delimiter=",", skip_header=1)
data = data.round()
data[:10]
```

```python
# solution
pos = np.lexsort((-data[:, 4], data[:, 2])) 
data[pos, :]
```
