**Some Useful Functions that might come into play**<br><ul>
<li>Digitize: bin a numeric variable into predefined categories.</li>
<li>Clip: clip a numeric variable to a defined min. and max. values</li>
<li>Histogram: produce the bins and the frequencies for a numeric variable. i.e: Elements of histogram</li>
</ul>

```python
import numpy as np

arr = np.genfromtxt('Datasets/iris.csv', delimiter=',', skip_header=1)

dt = np.dtype({'names': ['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width'],
               'formats': [float, float, float, float]})

arr.dtype = dt

arr[:10]
```

```python
sepal_length = arr['Sepal.Length'].reshape(-1, )
sepal_length
```

**np.digitize** <br>
Bin a numeric variable to predefined categories. Numeric -> Categorical

```python
np.digitize(sepal_length, bins=[0,5,6,7,10])
```

**Clip**

```python
np.clip(sepal_length,5 ,7)
```

**Histogram**

```python
np.histogram(sepal_length, bins=10)
```

```python
import numpy as np

np.random.seed(100)

arr = np.random.normal(30, 10, 100).round(2)

arr
```

```python
minimum = np.quantile(arr, 0.10)
maximum = np.quantile(arr, 0.90)

print(minimum,", ", maximum)
np.clip(arr, a_min=minimum, a_max=maximum)
```
