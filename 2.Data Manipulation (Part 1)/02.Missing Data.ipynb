# **Missing Data**

In Python, anything that is missing is represented as `None`.

In NumPy, since we are dealing specifically with data we use a more specific notation for missing values: `np.nan`. Infinity is represented as `np.inf`.

```python
import numpy as np
```

```python
np.nan
```

Be Careful when doing computation with missing values (`np.nan`)

```python
np.nan == np.nan  # False, so don't use comparison of nan.
```

```python
np.nan in [np.nan] # works well
```

```python
np.nan is np.nan  # works well
```

**Import Data with missing value**

```python
data = np.genfromtxt('Datasets/data_miss.txt', delimiter='\t')
data
```

**Check for missing data in array**

```python
np.isnan(data)
```

**Check for infinity**

```python
np.isinf(data)
```

**Missing or Infinity**

```python
np.isnan(data) | np.isinf(data)
```

**Fill in the missing or infinite with some value**

```python
data[np.isnan(data) | np.isinf(data)] = 0
```
